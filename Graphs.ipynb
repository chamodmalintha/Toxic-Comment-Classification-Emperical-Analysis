```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
```

```python
#Dataset 01 - Results

data = np.array([[91, 91], [91, 93], [91, 93], [93, 93], [75, 77], [90, 90]])
data_std = np.array([[1, 2], [1, 2], [1, 2], [1, 2], [1, 2], [1, 2]])    

length = len(data)
x_labels = ['NB', 'SVM-linear', 'SVM-rbf', 'LR', 'KNN', 'RF']

# Set plot parameters
fig, ax = plt.subplots()
width = 0.2 # width of bar
x = np.arange(length)

ax.bar(x, data[:,0], width, color='b', label='Bow')
ax.bar(x + width, data[:,1], width, color='g', label='Tf-idf')

ax.set_ylabel('Accuracy')
ax.set_ylim(0,100)
ax.set_xticks(x + width + width/2)
ax.set_xticklabels(x_labels)
ax.set_xlabel('Models')
ax.set_title('Dataset 01 - Results')
ax.legend()
plt.grid(True, 'major', 'y', ls='--', lw=.5, c='k', alpha=.3)

fig.tight_layout()
plt.show()
```
